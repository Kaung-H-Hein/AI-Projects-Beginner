Importing the Dependencies

```python
import numpy as np # to make numpy arrays (helpful in processing)
import pandas as pd # to create dataframes (to put data in nice structured tabels)
from sklearn.preprocessing import StandardScaler # to starndize the data
from sklearn.model_selection import train_test_split # to split into training data & testing data
from sklearn import svm # support vector machine from sklearn(model)
from sklearn.metrics import accuracy_score  #to get accuracy_score for evaluating our model
```

Data Collection and Analysis

diabetes.csv

```python
# loading the diabetes dataset and change it into a pandas DataFrame
diabetes_dataset =  pd.read_csv('/content/diabetes.csv')
```

```python
# printing the first 5 rows of the dataset
diabetes_dataset.head(5)
```

```python
# number of rows and colums in the dataset
diabetes_dataset.shape
```

```python
# getting the statistical measures of the data
diabetes_dataset.describe()
```

```python
# counting the outcome data (0 & 1)
diabetes_dataset['Outcome'].value_counts()
```

0 = Non-Diabetic,
1 = Diabetic

```python
# grouped the data based on 0 & 1, found the mean values for patient with diabetes and non-diabetes
# differences in glucose & insulin. (diabetic person has more level)
diabetes_dataset.groupby('Outcome').mean()
```

```python
# separating the data and labels
X = diabetes_dataset.drop(columns = 'Outcome', axis = 1)
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Data Standardization

Standardize the data to get a common range as the data are in differenet ranges

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

Train / Test Split

split 20% (0.2) into train data.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the Model

loaded our model support vector machine classifier into vairable classifier.

```python
classifier = svm.SVC(kernel='linear')
```

```python
# training the support vector Machine Classifier
classifier.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on the train data
X_train_prediction = classifier.predict (X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print ('Accuracy score of the train data : ' , training_data_accuracy)
```

```python
# accuracy score on the test data
X_test_prediction = classifier.predict (X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print ('Accuracy score of the test data : ' , testing_data_accuracy)
```

Making a Prediction System

- 4,110,92,0,0,37.6,0.191,30 = 0 (Non-Diabetic)
- 5,166,72,19,175,25.8,0.587,51 = 1 (Diabetic)

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
# reason of reshaping the array: our model is trainied on 768 examples with 8 colums, but in this case we are using only 1 data point.
# if we don't reshape the array, the model expects 768 data points and make the model will get confuse
# thus with reshapping the data, the model will predict only for one data

input_data_reshaped = input_data_as_numpy_array.reshape(1, -1) # this is parament that tells the models we r not giving 768 examples, we''ll only predict for only 1 datapoint
# we can't use raw data instead we need to standardize the data as we standardize our training data above, if we use raw data, the predicted outcome will be incorrect

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
